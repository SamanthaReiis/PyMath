# Geometria Analítica com Python


### 1. Rotações de curva

 - Vamos iniciar fazendo o gráfico da função cosseno

```python
# Importando bibliotecas
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Parâmetro
t = np.linspace(0, 2 * np.pi, 100)

# Coordenada em função do parâmetro
x = t
y = np.cos(t)
```

Observe que x=t e y=cos(t), logo y = cos(x)

```python
# Construindo o gráfico
plt.plot(x,y)
plt.title('Função cosseno', fontsize=16, color='g')

# Plotando o gráfico
plt.show()
```

Para rotacionar o gráfico acima, iremos primeiro conheceremos a matriz de rotação:

$$ \begin{pmatrix} 
  cos(\theta) & -sin(\theta) \\ 
  sin(\theta) & cos(\theta) \\
  \end{pmatrix}   $$

As variações antigas de x e y em função de t eram x=t e y=cos(t). Multiplicaremos essas variações pela matriz de rotação:
                   
$$ \begin{pmatrix}cos(\theta) & -sin(\theta) \\ sin(\theta) & cos(\theta) \\\end{pmatrix} \  \begin{pmatrix}t \\ cos(\theta) \\\end{pmatrix}$$ 

                                         

Após realizamos a multiplicação obtemos:
\begin{pmatrix}tcos(\theta) - sin(\theta)cos(t)\\ tsin(\theta) + cos(\theta)cos(t)\\\end{pmatrix}

As novas variações de x e y após a rotação, no sentido anti-horário, por um ângulo θ. Isto é;

$$ x = t\cos(\theta) - \sin(\theta)\cos(t) $$
$$ y = t\sin(\theta) + \cos(\theta)\cos(t) $$

Agora, na prática, faremos uma rotação, no sentido anti-horário, de 45º da curva y=cos(x).

```python
# Parâmetro
t= np.linspace(0, 2*np.pi, 100)

# Ângulo
theta = np.pi/4

# Coordenadas em função do parâmetro
x = np.cos(theta) * t - np.sin(theta) * np.cos(t)
y= np.sin(theta) * t + np.cos(theta) * np.cos(t)

# Construindo o gráfico
plt.plot(x, y)
plt.title('Rotação de 45º da curva y = cos(x)')

# Ajustando os eixos
plt.axis('equal')

# Plotando o gráfico
plt.show()
```

##### 1.1 Elipse

```python
# Parâmetro
t = np.linspace(0, 2*np.pi, 100)

# Equações paramétricas da Elipse
x = 4 * np.cos(t)
y = 2 * np.sin(t)

# Construindo o gráfico
plt.plot(x, y, color='r')
plt.title('Elipse', color='r', fontsize=18)

# Plotando o gráfico
plt.show()
```

Vamos rotaciona a elipse num ângulo de 45º no sentido anti-horário, no entanto, observe que escreveremos o código de maneira conveniente para os próximos passos.

```python
# Parâmetro
t = np.linspace(0, 2*np.pi, 100)

# Equações paramétricas da Elipse
x = 4 * np.cos(t)
y = 2 * np.sin(t)

# Angulo
theta = np.pi/4

# Coordenadas em função do parâmetro
x_theta = np.cos(theta) * x - np.sin(theta) * y
y_theta = np.sin(theta) * x + np.cos(theta) * y

# Fazendo o gráfico
plt.plot(x_theta, y_theta, color='orange')
plt.title('Elipse rotacionada de um ângulo de 45', color='orange', fontsize=18)

# Ajustando o eixos
plt.axis('equal')

# Plotando o gráfico
plt.show()

```

```python
# Parâmetro
t = np.linspace(0, 2*np.pi, 100)

# Equações paramétricas da Elipse
x = 4 * np.cos(t)
y = 2 * np.sin(t)

# Angulo
for n in range(1,9):
    theta = n * (np.pi/4) 
    # Coordenadas em função do parâmetro
    x_theta = np.cos(theta) * x - np.sin(theta) * y
    y_theta = np.sin(theta) * x + np.cos(theta) * y
    # Fazendo o gráfico
    plt.plot(x_theta, y_theta, color='magenta')
    plt.title(' Elipses rotacionadas', color='purple', fontsize=18)

# Ajustando o eixos
plt.axis('equal')

# Plotando o gráfico
plt.show()

```

Agora diminuimos o ângulo e aumentando o número de rotações

```python
# Tamanho da janela devisuaização da figura
plt.figure(figsize=(16,16))

#Parâmetro
t = np.linspace(0, 2*np.pi, 100)

# Equações paramétricas da Elipse
x = 4 * np.cos(t)
y = 2 * np.sin(t)

# Angulo
for n in range(1, 180):
    theta = n * (np.pi/90) 
    # Coordenadas em função do parâmetro
    x_theta = np.cos(theta) * x - np.sin(theta) * y
    y_theta = np.sin(theta) * x + np.cos(theta) * y
    # Fazendo o gráfico
    plt.plot(x_theta, y_theta, color='magenta')
    plt.title(' Elipses rotacionadas', color='purple', fontsize=18)

# Ajustando o eixos
plt.axis('equal')

# Plotando o gráfico
plt.show()
```

##### 1.2 Parábola


```python
 # dados do círculo
r =  4.5
 
# Parâmetro
s = np.linspace(np.pi * (-4.0), np.pi * 4.0, 100)     
x_c =  r * np.cos(s)                   # calculando X e Y
y_c = r * np.sin(s)

fig, ax = plt.subplots()


# Parâmetro
t = np.linspace(-4, 4, 100)

# Equações paramétricas
x = t 
y= (1/8) * t ** 2

# Angulo
for n in range(1,25):
    theta = n * (np.pi/10) 
    # Coordenadas em função do parâmetro
    x_theta = np.cos(theta) * x - np.sin(theta) * y
    y_theta = np.sin(theta) * x + np.cos(theta) * y
    # Fazendo o gráfico
    plt.plot(x_theta, y_theta, color='magenta', alpha=1)
    plt.title(' Parábolas rotacionadas dentro de um círculo', color='purple', fontsize=14)
    
    #ax.plot(x_c, y_c, 'g-')                    
    ax.fill(x_c, y_c, alpha=0.3, color='black')           # usando fill no lugar do plot para preencher
    ax.plot([1], [2], color='black' )                  # mostrando o centro



# Ajustando o eixos
plt.axis('equal')

# Plotando o gráfico
plt.show()
```
